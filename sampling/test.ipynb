```python
import os
import torch
import re
from safetensors.torch import load_file
from copy import deepcopy

BASE_DIR = "/teamspace/studios/this_studio/.porting"

def load_flux_fill():
    model_path = os.path.join(BASE_DIR, "models", "unet", "flux1-fill-dev-fp8.safetensors")
    print(f"Loading base model: {os.path.basename(model_path)}\n")
    return load_file(model_path)

def translate_lora_key_to_model_key(lora_key):
    return re.sub(r'(_)(\d+)', r'.\2', lora_key)

if __name__ == "__main__":
    LORA_FILENAME = "comfyui_subject_lora16.safetensors"
    LORA_STRENGTH = 1.0

    base_weights = load_flux_fill()
    lora_path = os.path.join(BASE_DIR, "models", "loras", LORA_FILENAME)
    if not os.path.exists(lora_path):
        raise FileNotFoundError(f"LoRA not found at {lora_path}")
    
    print(f"Loading LoRA: {LORA_FILENAME}\n")
    lora_weights = load_file(lora_path)
    
    weights_with_loras = deepcopy(base_weights)

    print("--- Starting Layer-by-Layer Verification ---\n")
    
    matched_keys = 0
    total_lora_keys = 0

    # Isolate all relevant 'lora_up' keys from the LoRA file
    lora_up_keys = sorted([k for k in lora_weights.keys() if "lora_up.weight" in k and k.startswith("lora_unet_")])
    total_lora_keys = len(lora_up_keys)

    for up_key in lora_up_keys:
        down_key = up_key.replace("lora_up.weight", "lora_down.weight")
        alpha_key = up_key.replace("lora_up.weight", "alpha")

        if down_key not in lora_weights:
            continue

        # --- 1. Translate the key ---
        lora_base_key = up_key.split(".lora_up.weight")[0]
        lora_style_key = lora_base_key[len("lora_unet_"):] + ".weight"
        model_key = translate_lora_key_to_model_key(lora_style_key)
        
        print(f"Processing LoRA layer: '{lora_style_key}'")

        if model_key in weights_with_loras:
            matched_keys += 1
            print(f"  - Match found in base model: '{model_key}'")

            # --- 2. Get original value (Before) ---
            original_value = torch.sum(weights_with_loras[model_key]).item()
            print(f"    - Before: {original_value:.4f}")

            # --- 3. Apply the LoRA update ---
            lora_down = lora_weights[down_key]
            lora_up = lora_weights[up_key]
            
            # Get alpha, default to rank if not present
            rank = lora_down.shape[0]
            alpha = lora_weights.get(alpha_key, torch.tensor(rank)).item()
            
            # Calculate the update, including the alpha scaling
            scale = (alpha / rank) if rank > 0 else 1.0
            update = (lora_up.float() @ lora_down.float()) * scale * LORA_STRENGTH
            
            weights_with_loras[model_key] += update.to(weights_with_loras[model_key].dtype)

            # --- 4. Get new value (After) ---
            modified_value = torch.sum(weights_with_loras[model_key]).item()
            print(f"    - After:  {modified_value:.4f}")
            print("-" * 20)

        else:
            print(f"  - WARNING: No match found in base model for this layer.")
            print("-" * 20)

    print("\n--- Verification Summary ---")
    print(f"Successfully matched and applied {matched_keys} of {total_lora_keys} UNet LoRA layers.")
    print("--- Script Finished ---\n")
```
